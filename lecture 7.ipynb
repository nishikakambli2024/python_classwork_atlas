```python
report_card_dictionary = {"student_names": ["Nishika","Aakriti","Aayushi","Peony","Riya"],
                        "physics_grades": [34, 45, 56, 78, 100],
                        "chemistry_grades": [100, 78, 0, 45, 98],
                        "math_grades": [56, 76, 34, 23, 97],
                        "biology_grades": [45, 56, 34, 67, 78]
                        }
report_card_dictionary
```

```python
#setup
import pandas as pd
```

```python
#making the report card dictionary into a dataframe
report_card_df = pd.DataFrame(report_card_dictionary)
report_card_df
```

```python
#first 3 rows in the report card dataframe 
report_card_df.head(3)
```

```python
# 3 random rows in the report card dataframe
report_card_df.sample(3)
```

```python
#describing all the numeric column in the report card dataframe 
report_card_df.describe()
```

#insights
- average chemistry grade? 64.2
- average biology grade? 56.0
- lowest math grade? 23.0
- highest physics grade? 100
- scores preesent in the math column? 5

```python
report_card_df["total_grades"] = report_card_df["physics_grades"]+ report_card_df["math_grades"]+ report_card_df["chemistry_grades"]+ report_card_df["biology_grades"]
report_card_df
```

```python
report_card_df["avg_grades"] = report_card_df["total_grades"]/4
report_card_df
```

```python
report_card_insights_df = report_card_df.describe().transpose()
report_card_insights_df
```

```python
#saving a csv file 
report_card_insights_df.to_csv("report card insights.csv")
```

REPORT CARD ALPHABET GRADES

```python
report_card_df
```

```python
report_card_df.loc[report_card_df["total_grades"]>=140, "Status"]= "Pass"
report_card_df.loc[report_card_df["total_grades"]<140, "Status"]= "Fail"
report_card_df
```

```python
#making an alphabet grade column 
report_card_df.loc[(report_card_df["avg_grades"] >=90), "Character_Grade"]= "A"
report_card_df.loc[(report_card_df["avg_grades"] >=70) & (report_card_df["avg_grades"]<90), "Character_grade"]= "B"
report_card_df.loc[(report_card_df["avg_grades"] >=50) & (report_card_df["avg_grades"]<70), "Character_grade"]= "C"
report_card_df.loc[(report_card_df["avg_grades"] >=35) & (report_card_df["avg_grades"]<50), "Character_grade"]= "D"
report_card_df.loc[(report_card_df["avg_grades"] <35), "Character_grade"]= "F"
```
