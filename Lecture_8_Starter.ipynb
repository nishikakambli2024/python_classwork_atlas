# Functions

```python
#set up 
import pandas as pd
```

```python
#get scores

def get_scores(report_card_dictionary, division_name):
    
    #convert the dictionary into dataframe 
    report_card_df = pd.DataFrame(report_card_dictionary)
    
    #add a coloumn called division (e.g A, B, C) 
    report_card_df["division"] = division_name
    
    #add a coloumn called total score (0 to 400)
    report_card_df["total_score"] = report_card_df["math_scores"] + report_card_df["chemistry_scores"] + report_card_df["physics_scores"] + report_card_df["information_technology_scores"]
    
    #add a column called average score (0 to 100)
    report_card_df["average_scores"] = report_card_df["total_score"]/4
    
    #add a column called status (pass or fail)
    report_card_df.loc[report_card_df["average_scores"] >=35, "status"] = "Pass"
    report_card_df.loc[report_card_df["average_scores"] <35, "status"] = "fail"
    
    #add a column called character grade ( A, B, C, D, F)
    report_card_df.loc[report_card_df["average_scores"] >=90, "character_grade"] = "A"
    report_card_df.loc[(report_card_df["average_scores"] >=70) & (report_card_df["average_scores"] <90), "character_grade"] = "B"
    report_card_df.loc[(report_card_df["average_scores"] >=50) & (report_card_df["average_scores"] <70), "character_grade"] = "C"
    report_card_df.loc[(report_card_df["average_scores"] >=35) & (report_card_df["average_scores"] <50), "character_grade"] = "D"
    report_card_df.loc[report_card_df["average_scores"] <35, "character_grade"] = "F"
    
    #return the dataframe 
    return report_card_df
```

```python
division_a_report_card_df = get_scores(div_a_scores_dictionary, "A")
division_a_report_card_df
```

GET EMAILS FUNCTION

```python
#

def get_emails(emails_dictionary, division_name):
    
    #converting the dictionary into a dataframe
    emails_df = pd.DataFrame(emails_dictionary)
    
    #creating a column called division 
    emails_df["division"] = division_name
    
    #converting all string functions in the student email column to lower case
    emails_df["student_email"] = emails_df["student_email"].str.lower()
    
    #return the dataframe
    return emails_df
    
```

```python
div_a_emails_df = get_emails(div_a_emails_dictionary, "A")
div_a_emails_df
```

GET SCORES AND EMAIL FUNCTION

```python
def get_scores_and_emails_combined(report_card_df, emails_df):
    report_card_with_emails_df = pd.merge(report_card_df, emails_df, how = "left", on = ["student_name", "division"])
    return report_card_with_emails_df
```

```python
get_scores_and_emails_combined(division_a_report_card_df, div_a_emails_df)
```

GET ALL DIVISIONS SCORESHEET 

```python
div_a_scores_df = get_scores(div_a_scores_dictionary, "A")
div_a_emails_df = get_emails(div_a_emails_dictionary, "A")
div_a_combined_df = get_scores_and_emails_combined(div_a_scores_df, div_a_emails_df)
div_a_combined_df
```

```python
def get_all_divisions_scoresheet(div_a_df, div_b_df, div_c_df):
    all_divisions_df = pd.concat([div_a_df, div_b_df, div_c_df], ignore_index = True)
    return all_divisions_df
```
